Merging,Joining,and Concatenating

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df3
```

Concat
* Glue the Dataframe together
* Use Pandas Methods

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

Merging
* It works on Columns. Common Columns from both Dataframe
* Use Pandas Method

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K5'],
                     'A': ['A0', 'A1', 'A2', 'A3','A5'],
                     'B': ['B0', 'B1', 'B2', 'B3','B5']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','k4'],
                          'C': ['C0', 'C1', 'C2', 'C3','C4'],
                          'D': ['D0', 'D1', 'D2', 'D3','D4']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='left',on='key')
```

Join
* It works on Index/ Rows (Common Index/ row from both Dataframe)
* Dataframe Method




```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right,how='inner')
```
