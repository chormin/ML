**distplot / displot / histogram**
* This plot show the **distribution** of a **univarient** set of observation
* Bin width = (Max-Min) / number of Bins


```python
tips.head()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], bins=3)
```

Joinplot
* Combination of two displot

```python
sns.jointplot(data=tips,x='total_bill',y='tip', hue='time')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip', kind= 'kde' )
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip', kind= 'hex', )
```

pairplot

```python
tips.head()
```

```python
sns.pairplot(tips,hue='sex')
```

KDE Plot

```python
sns.kdeplot(tips['total_bill'])
```

Countplot
* Categorical Plot
* Used for **Univarient** Categorical Data

```python
tips['sex'].value_counts()
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.countplot(data=tips, x='sex', hue='time')
```

Bar plot

```python
import numpy as np
```

```python
sns.barplot(data=tips,x='sex',y='total_bill', estimator=np.std )
```

boxplot

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

Violinplot

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=tips,x='day',y='total_bill')
```

stripplot

* Draw scatterplot for categorical data

```python
tips.head()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(data=tips,x='day',y='total_bill', hue='sex')
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex')
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=tips,x='day',y='total_bill')
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex',color='black')
```

Matrix Plot

* heatmap
* clustermap

```python
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

```python
flight_pivot = flight.pivot_table(values='passengers', index='month', columns='year' )
```

```python
flight_pivot
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_pivot,cmap='coolwarm',linewidth=1,linecolor='white',annot=True,fmt='g')
```

```python
sns.clustermap(flight_pivot,cmap='coolwarm',linewidth=1,linecolor='white',annot=True,fmt='g')
```
